```python
import pandas as pd
```

```python
df = pd.read_csv("../data/insurance_cleaned.csv")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = df.corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap")
plt.show()
```

```python
corr_target = corr['charges'].sort_values(ascending=False)

corr_target
```

```python
corr_target.drop('charges').plot(
    kind='bar',
    figsize=(8,4),
    title='Feature Correlation with Charges'
)
plt.show()
```
