```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/insurance.csv")
df.head()
```

### Check the Normality of the Target Variable.

```python
plt.figure(figsize=(6,4))
sns.histplot(df['charges'], kde=True)
plt.title("Distribution of Medical Insurance Charges")
plt.show()
```

*Target variable is highly Right Skewed,possibility of outlier*

### Box Plot for Analysis

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Insurance Charges by Smoking Status")
plt.show()
```

*Smokers Features affect Insurance charge the most.*

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='children', y='charges', data=df)
plt.title("Charges vs Number of Children")
plt.show()
```

*Number of children<3 has more outlier. lower the number of childer also increase in charges.*

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='region', y='charges', data=df)
plt.title("Charges by Region")
plt.show()
```

### Scatter Plots

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='age', y='charges', hue='smoker', data=df)
plt.title("Charges vs Age with smoker")
plt.show()
```

*Charges increase with age*. **Older smokers show extreme costs.**

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='bmi', y='charges', hue='smoker', data=df)
plt.title("Charges vs BMI (Colored by Smoking Status)")
plt.show()
```
